```python
# Class representation of water tables

class WaterSourceType:
    def __init__(self, id, last_update, last_update_by, uuid, name, notes):
        self.id = id
        self.last_update = last_update
        self.last_update_by = last_update_by
        self.uuid = uuid
        self.name = name
        self.notes = notes


class WaterPointType:
    def __init__(self, id, last_update, last_update_by, uuid, name, notes):
        self.id = id
        self.last_update = last_update
        self.last_update_by = last_update_by
        self.uuid = uuid
        self.name = name
        self.notes = notes

class WaterPoint:
    def __init__(self, id, last_update, last_update_by, uuid, geometry, notes, image, water_source_type_uuid, water_point_type_uuid):
        self.id = id
        self.last_update = last_update
        self.last_update_by = last_update_by
        self.uuid = uuid
        self.geometry = geometry
        self.notes = notes
        self.image = image
        self.water_source_type_uuid = water_source_type_uuid
        self.water_point_type_uuid = water_point_type_uuid


class WaterLineType:
    def __init__(self, id, name, uuid, last_update, last_update_by):
        self.id = id
        self.name = name
        self.uuid = uuid
        self.last_update = last_update
        self.last_update_by = last_update_by

class WaterLine:
    def __init__(self, id, notes, last_update, last_update_by, uuid, image, water_line_type_uuid, water_source_uuid):
        self.id = id
        self.notes = notes
        self.last_update = last_update
        self.last_update_by = last_update_by
        self.uuid = uuid
        self.image = image
        self.water_line_type_uuid = water_line_type_uuid
        self.water_source_uuid = water_source_uuid
```
