```python
# Point validator. This verifies whether the entered point lies within
# the bounds of the latitudes and longitudes of the world.
```

```python
# Point validator

class PointValidator:
    def __init__(self):
        self.history = []
    
    def validate_point(self, lat, lon):
    
        valid = True
        reason = f'All seems in order ;).'

        if lat < -90 or lat > 90:
            valid = False
            reason = f'The latitude does not exist, it must be between -90 and 90.'
        
        if lon < -180 or lon > 180:
            valid = False
            reason = f'The longitude is invalid, it does not lie between -180 and 180.'
        
        if (lat < -180 or lat > 180) and (lon < -180 or lon > 180):
            valid = False
            reason = f'Both latitude and longitude are invalid.'

        print(f'Reason: {reason}')
        print(f'Coordinate Valid: {valid}')
        self.history.append((lat, lon, valid, reason))
            
    
    def check_history(self):
        print(f"Total points validated: {len(self.history)}")
        for i in self.history:
            print(i)


    def save_history(self):
        #Import datetime
        from datetime import datetime
        now = datetime.now()
    

        with open(f'point-{now}', 'x') as f:
            for i in self.history:
                f.write(str(i))
        
        print(f'History saved to file')

    def readfile(self):
        #Receive user input to open/read file
        filesname = input("Enter filename. Specify path if file located in different location to this file")
        with open(filesname, 'r') as f:
            data = f.read()
        
        #Add read info to history list
        self.history.append(data)

        print(data) 

    
    def run(self):
        while True:
            print(""" 
        COORDINATE POINT VALIDATOR

        OPTIONS:
        1: Validate a point
        2: Show history
        3: Save history
        4: Read file
        5: End validator
        
        """)
            command = int(input("Choose an option: "))

        
            if command == 1: #Validate
                try:
                    lat = float(input("Enter latitude: "))
                    lon = float(input("Enter longitude: "))
                    
                except (ValueError):
                    print("Please enter an integer number")
                else:
                    self.validate_point(lat, lon)

            elif command == 2: #Show history
                self.check_history()
            elif command == 3: #Save history
                    self.save_history()
            elif command == 4:
                try:
                    self.readfile()
                except FileNotFoundError:
                    print("File not found. Check path syntax, spelling or file extension")
            elif command == 5: #End validator
                break
            else:
                print("Command does not exist")

# Create instance of class, then run it.
point_validator = PointValidator()
point_validator.run()
```
